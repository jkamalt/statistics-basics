```python
from math import factorial
```

```python
# Вычисляет число сочетаний k из из n элементов
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# Вычисляет число размещений k из из n элементов
def arrangements(n, k):
    return int(factorial(n) / factorial(n - k))
```

### Задание 1

*Из колоды в 52 карты извлекаются случайным образом 4 карты.*

*a) Найти вероятность того, что все карты – крести.*

*б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.*

#### Пример 1
*Найти вероятность того, что все карты – крести.*

Для вычисления вероятности нужно число исходов $m$, благоприятствующих событию, разделить на общее число исходов $n$.

$$P = \frac{m}{n}.$$

Число исходов $m$, благоприятствующих событию - количество сочетаний 4 карт из 52, состоящих только из крести. Всего колода из 52 карт содержит 13 карт крести. Тогда количество сочетаний 4 карт из 13 карт крести:

$$m = C_{13}^4 = \frac{13!}{4! (13 - 4)!} = 715.$$

Общее число исходов $n$ - все возможные сочетания 4 карт из 52:

$$n = C_{52}^4 = \frac{52!}{4! (52 - 4)!} = 270725.$$

Тогда вероятность того, что все карты - крести

$$P = \frac{m}{n} = \frac{715}{270725} \approx 0.00264.$$

**Ответ:** $P \approx 0.00264$.

```python
m = combinations(13, 4)
n = combinations(52, 4)
P = m / n
print(f'm = {m}\nn = {n}')
print(f'P = {P}')
```

#### Пример 2
*Найти вероятность, что среди 4-х карт окажется хотя бы один туз.*

Для вычисления вероятности нужно число исходов $m$, благоприятствующих событию, разделить на общее число исходов $n$.

$$P = \frac{m}{n}.$$

Число исходов $m$, благоприятствующих событию - количество сочетаний 4 карт из 52, где будет хотя бы 1 туз, т.е. 1, 2, 3 или 4 туза. Рассмотрим отдельно 4 события:

$A$: 1 туз и 3 обычные,

$B$: 2 туза и 2 обычные,

$C$: 3 туза и 1 обычные,

$D$: 4 туза.

Это несовместные события, поэтому искомая вероятность равна сумме вероятностей отдельных событий:

$$P = P_A + P_B + P_C + P_D = \frac{m_A + m_B + m_C + m_D}{n},$$

т.к. для всех событий общее число исходов $n$ одинаковое - все возможные сочетания 4 карт из 52:

$$n = C_{52}^4 = 270725.$$

Колода из 52 карт: 4 туза и 48 обычных карт. 

1) Найдем число исходов $m_A$, благоприятствующих событию $A$ - 1 туз и 3 обычные. 

1 туз из 4 тузов можно выбрать 4 способами: $C_{4}^1 = 4$.

Число сочетаний 3 карт из 48 обычных карт: $C_{48}^3 = 17296$.

Тогда

$$m_A = C_{4}^1 \cdot C_{48}^3 = 69184.$$

2) Найдем число исходов $m_B$, благоприятствующих событию $B$ - 2 туза и 2 обычные. 

Число сочетаний 2 тузов из 4 тузов: $C_{4}^2 = 6$.

Число сочетаний 2 карт из 48 обычных карт: $C_{48}^2 = 1128$.

Тогда

$$m_B = C_{4}^2 \cdot C_{48}^2 = 6768.$$

3) Найдем число исходов $m_C$, благоприятствующих событию $C$ - 3 туза и 1 обычные. 

Число сочетаний 3 тузов из 4 тузов: $C_{4}^3 = 4$.

Одну карту из 48 обычных карт можно выбрать 48 способами: $C_{48}^1 = 48$.

Тогда

$$m_C = C_{4}^3 \cdot C_{48}^1 = 192.$$

4) Найдем число исходов $m_D$, благоприятствующих событию $D$ - 4 туза. 

4 туза из 4 тузов можно выбрать 1 способом: $C_{4}^4 = 1$.

Тогда

$$m_D = C_{4}^4 = 1.$$

Тогда вероятность того, что среди 4-х карт окажется хотя бы один туз

$$P = \frac{m_A + m_B + m_C + m_D}{n} = \frac{69184 + 6768 + 192 + 1}{270725} \approx 0.281.$$

**Ответ:** $P \approx 0.281$.

```python
n = combinations(52, 4)
print(f'n = {n}\n')

mA_t = combinations(4, 1)
mA_n = combinations(48, 3)
mA = mA_t * mA_n
print(f'mA_t = {mA_t}\nmA_n = {mA_n}')
print(f'mA = {mA}\n')

mB_t = combinations(4, 2)
mB_n = combinations(48, 2)
mB = mB_t * mB_n
print(f'mB_t = {mB_t}\nmB_n = {mB_n}')
print(f'mB = {mB}\n')

mC_t = combinations(4, 3)
mC_n = combinations(48, 1)
mC = mC_t * mC_n
print(f'mC_t = {mC_t}\nmC_n = {mC_n}')
print(f'mC = {mC}\n')

mD = combinations(4, 4)
print(f'mD = {mD}\n')

P = (mA + mB + mC + mD) / n
print(f'P = {P}')
```

### Задание 2

*На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно.*

*Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?*

Для вычисления вероятности нужно число исходов $m$, благоприятствующих событию, разделить на общее число исходов $n$.

$$P = \frac{m}{n}.$$

Число исходов $m$, благоприятствующих событию - 1 правильный код.

$$m = 1.$$

Общее число исходов $n$ - все возможные сочетания 3 цифр из 10, т.к. кнопки нажимаются одновременно:

$$n = C_{10}^3 = \frac{10!}{3! (10 - 3)!} = 120.$$

Тогда вероятность открыть дверь с первой попытки:

$$P = \frac{m}{n} = \frac{1}{120} \approx 0.0083.$$

**Ответ:** $P \approx 0.0083$.

```python
m = 1
n = combinations(10, 3)
P = m / n
print(f'm = {m}\nn = {n}')
print(f'P = {P}')
```

### Задание 3

*В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали.*

*Какова вероятность того, что все извлеченные детали окрашены?*

Для вычисления вероятности нужно число исходов $m$, благоприятствующих событию, разделить на общее число исходов $n$.

$$P = \frac{m}{n}.$$

Число исходов $m$, благоприятствующих событию - количество сочетаний 3 деталей из 9 окрашенных:

$$m = C_{9}^3 = \frac{9!}{3! (9 - 3)!} = 84.$$

Общее число исходов $n$ - все возможные сочетания 3 деталей из 15 деталей:

$$n = C_{15}^3 = \frac{15!}{3! (15 - 3)!} = 455.$$

Тогда вероятность того, что все извлеченные детали окрашены

$$P = \frac{m}{n} = \frac{84}{455} \approx 0.185.$$

**Ответ:** $P \approx 0.185$.

```python
m = combinations(9, 3)
n = combinations(15, 3)
P = m / n
print(f'm = {m}\nn = {n}')
print(f'P = {P}')
```
